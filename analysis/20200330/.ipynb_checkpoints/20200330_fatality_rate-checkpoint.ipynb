# Organizing data 

```python
# Helpers
def convert_col(x):
    tmp = []
    for item in x:
        if type(item) is str:
            tmp.append(float(item.split()[0]))
        else:
            tmp.append(item)
    return tmp

def convert_col_custom(x, func):
    tmp = []
    for item in x:
        if type(item) is str:
            tmp.append(func(item))
        else:
            tmp.append(item)
    return tmp
```

```python
# Convert date to standard date code
def convert_date(date):
    tmp = {"Mar":"3", "Feb":"2"}
    year = "20"
    month = date.split(",")[0].split()[0]
    day = date.split(",")[0].split()[1]
    return tmp[month] + "/" + day + "/" + year
data_tests["Date"] = [convert_date(item) for item in data_tests["Date"].values]
```

```python
data_tests.head()
```

Cleaning the data. Cf: https://ourworldindata.org/coronavirus-testing-source-data

1. Australia: only the entry with the index "Australia" is retained. This is supposed to be a sum of per-state numbers.
2. Canada is removed, because of caveats mentioned in the link above. 
3. China - Guangdong is removed, since we don't have other data types in the provincial level of China. 
4. United States: only the entry "United States" is retained. The entry "United States – CDC samples tested" is removed.
5. Kuwait will be removed since there is no consistent data. 

```python
retained_indices = []
for item in data_tests.index:
    if item.startswith("Australia") and "-" in item:
        continue
    if item.startswith("Canada"):
        continue
    if item == "China - Guangdong":
        continue
    if item == "United States - CDC samples tested":
        continue
    if item == "Kuwait":
        continue
    retained_indices.append(item)
print(len(retained_indices))
```

## My data 

```python
# Helpers
def collect_UN_data(my_data, UN_data, series, colname, 
                    colname_country="Region/Country/Area",
                    colname_year="Year",
                    colname_series="Series",
                    colname_value="Value",
                    country_ambiguity=None):
    data = []
    for index, row in my_data.iterrows():
        if country_ambiguity is not None:
            if index in list(country_ambiguity.keys()):
                country = country_ambiguity[index]
            else:
                country = index
        else:
            country = index
        df = UN_data[UN_data[colname_country] == country]
        if df.shape[0] == 0:
            print("!!! No data: " + index)
            data.append(np.nan)
        else:
            df = df[(df[colname_series] == series)]
            years = df[colname_year]
            max_year = np.max(years)
            data.append(df[(df[colname_year] == max_year)][colname_value].values[0])
            print(index, max_year)
    print(len(data))
    my_data[colname] = data
    return my_data

def collect_UN_data_fixed_year(my_data, UN_data, year, series, colname,
                               colname_country="Region/Country/Area",
                               colname_year="Year",
                               colname_series="Series",
                               colname_value="Value"):
    data = []
    for index, row in my_data.iterrows():
        df = UN_data[UN_data[colname_country] == index]
        if df.shape[0] == 0:
            print("!!! No data: " + index)
            data.append(np.nan)
        else:
            df = df[(df[colname_year] == year)]
            if df.shape[0] == 0:
                print(index + ": " + str(year) + " data not available")
                data.append(np.nan)
            else:
                try:
                    tmp = df[(df[colname_series] == series)][colname_value].values[0]
                except:
                    print(index + ": " + str(year) + " data not available")
                    data.append(np.nan)
                else:
                    data.append(tmp)
    print(len(data))
    my_data[colname] = data
    return my_data

def collect_UN_data2(my_data, UN_data, colname, 
                     colname_country="Country or Area",
                     colname_year="Year(s)",
                     colname_value="Value",
                     country_ambiguity=None,
                     year_selector=lambda x: np.max(x)):
    data = []
    for index, row in my_data.iterrows():
        if country_ambiguity is not None:
            if index in list(country_ambiguity.keys()):
                country = country_ambiguity[index]
            else:
                country = index
        else:
            country = index
        df = UN_data[UN_data[colname_country] == country]
        if df.shape[0] == 0:
            print("!!! No data: " + index)
            data.append(np.nan)
        else:
            years = df[colname_year]
            max_year = year_selector(years)
            data.append(df[(df[colname_year] == max_year)][colname_value].values[0])
            print(index, max_year)
    print(len(data))
    my_data[colname] = data
    return my_data
```

```python
# Get corresponding confirmed cases and deaths
confirmed_cases = []
deaths = []
for index, row in my_data.iterrows():
    # Take care of mismatched country names
    # Countries that do not have data: Faeroe Islands and Palestine
    if index == "Czech Republic":
        country = "Czechia"
    elif index == "Taiwan":
        country = "Taiwan*"
    elif index == "South Korea":
        country = "Korea, South"
    elif index == "United States":
        country = "US"
    elif index == "Hong Kong":
        df1 = data_JHU_confirmed[data_JHU_confirmed["Province/State"] == index]
        df2 = data_JHU_deaths[data_JHU_deaths["Province/State"] == index]
        confirmed_cases.append(df1[row["Date"]].sum())
        deaths.append(df2[row["Date"]].sum())
        continue
    else:
        country = index
    df1 = data_JHU_confirmed[data_JHU_confirmed["Country/Region"] == country]
    df2 = data_JHU_deaths[data_JHU_deaths["Country/Region"] == country]
    if df1.shape[0] == 0 or df2.shape[0] == 0:
        print(index)
        confirmed_cases.append(np.nan)
        deaths.append(np.nan)
    else:
        confirmed_cases.append(df1[row["Date"]].sum())
        deaths.append(df2[row["Date"]].sum())
print(len(confirmed_cases))
print(len(deaths))
my_data["Confirmed cases"] = confirmed_cases
my_data["Deaths"] = deaths
print(my_data.shape)
# Remove Faerow Islands and Palestine
my_data = pd.DataFrame.copy(my_data.loc[[item for item in my_data.index.values if item not in ["Faeroe Islands", "Palestine"]]])
print(my_data.shape)
```

# Analysis 

## Filter data

```python
# Require at least 100 confirmed cases
data_df = my_data_cleaned_transformed.loc[my_data_cleaned[my_data_cleaned["Confirmed"] >= 100].index]
print(data_df.shape)
```

```python
# Do not consider CTPerMillion and CTOverload, otherwise number of data points is reduced by half.
# Also remove other covariates that are irrelevant for fatality rate.
data_df = data_df[[
    "Fatality", 
    "Tests", "TestsPerMillion", 
    "Population60Plus", "PopulationMedianAge", "PopulationDensity", 
    "GDPPerCapita", 
    "PhysiciansPer1000", "BedsPer10000", 
    "CO2PerCapita", 
    "Obesity", "Smoking", "RaisedBloodPressure", 
    "PhysicianOverload", "BedOverload",
    "GeographicRegion"
]]
print(data_df.shape)
```

```python
# Remove nan rows
retained_indices = []
for index in data_df.index:
    if np.sum(np.isfinite(data_df.loc[index].values[0:-1].astype(float))) != 15:
        continue
    else:
        retained_indices.append(index)
print(len(retained_indices))
data_df = pd.DataFrame.copy(data_df.loc[retained_indices])
print(data_df.shape)
```

## Marginal tests 

```python
mpl.rcParams['pdf.fonttype'] = 42
fig = plt.figure()
fig.set_size_inches(30, 10)
sns.set_context("notebook")
sns.set_style("ticks")
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12)

x = data_df["Fatality"].values
i = 0
for col in data_df.columns[1:13]:
    i += 1
    y = data_df[col].values
    r, p = stats.pearsonr(x, y)
    
    subfig = fig.add_subplot(2, 6, i)
    subfig.set_title(col + "\n r=%.3g, p=%.2g" %(r, p), fontsize = 14)
    subfig.spines['right'].set_visible(False)
    subfig.spines['top'].set_visible(False)
    subfig.spines['bottom'].set_color('k')
    subfig.spines['left'].set_color('k')
    for tick in subfig.get_xticklabels():
        tick.set_fontname("Arial")
    for tick in subfig.get_yticklabels():
        tick.set_fontname("Arial")
    subfig.set_xlabel("Fatality", fontsize=14)
    subfig.set_ylabel(col, fontsize=14)
    subfig.scatter(x, y)
    
plt.tight_layout()
plt.savefig("scatterplot_with_fatality.pdf", bbox_inches='tight')
```

```python
# Pairwise correlations
marginal_corr_pearson = []
for col in data_df.columns[1:]:
    r, p = stats.pearsonr(data_df["Fatality"], data_df[col])
    marginal_corr_pearson.append([r, p, col])
marginal_corr_pearson = sorted(marginal_corr_pearson, key=itemgetter(0), reverse=True)
```

```python
rs = np.array([item[0] for item in marginal_corr_pearson])
ps = np.array([item[1] for item in marginal_corr_pearson])
covariates = np.array([item[2] for item in marginal_corr_pearson])
_, padjusts, _, _ = sms.multipletests(ps, 0.2, method="fdr_bh")
```

## Linear regression 

```python
formula = "Fatality ~ " + " + ".join(data_df.columns.values.tolist()[1:])
res = smf.ols(formula=formula, data=data_df).fit()
```

```python
res.summary()
```

```python
df = res.conf_int()
df["tvalues"] = res.tvalues
df["pvalues"] = res.pvalues
df = df.loc[[item for item in df.index.values if item != "Intercept"]]
df = df.sort_values(by="pvalues")
```

```python
tvalues = df["tvalues"].values
pvalues = df["pvalues"].values
covariates = df.index.values

mpl.rcParams['pdf.fonttype'] = 42
fig = plt.figure()
fig.set_size_inches(20, 4)
sns.set_context("notebook")
sns.set_style("ticks")
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12)
subfig = fig.add_subplot(1, 1, 1)
subfig.set_title("t-values from linear regression of Fatality ~ .", fontsize = 14)
subfig.spines['right'].set_visible(False)
subfig.spines['top'].set_visible(False)
subfig.spines['bottom'].set_color('k')
subfig.spines['left'].set_color('k')
for tick in subfig.get_xticklabels():
    tick.set_fontname("Arial")
for tick in subfig.get_yticklabels():
    tick.set_fontname("Arial")
subfig.set_xlabel("", fontsize=14)
subfig.set_ylabel("t-value", fontsize=14)

norm = mpl.colors.Normalize(vmin=-4, vmax=4)
cmap = cm.coolwarm
m = cm.ScalarMappable(norm=norm, cmap=cmap)

plt.bar(np.arange(len(tvalues)), tvalues, color=m.to_rgba(tvalues))
plt.xticks(np.arange(len(tvalues)), covariates, rotation=45, ha="right", fontsize=16);

rects = subfig.patches
labels = ["p=%.3g" % p for p in pvalues]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    if height > 0:
        subfig.text(rect.get_x() + rect.get_width() / 2, height + 0.5, label,
                    ha='center', va='center')
    else:
        subfig.text(rect.get_x() + rect.get_width() / 2, height - 0.5, label,
                    ha='center', va='center')

plt.ylim(-4, 4.5)
plt.savefig("linear_regression.pdf", bbox_inches='tight')
```
